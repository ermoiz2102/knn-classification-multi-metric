# 🤖 K-Nearest Neighbors (KNN) Classification  

## Objective:
To implement KNN manually and using Scikit-learn, and compare predictions using multiple distance metrics.

## 📦 Importing Libraries  

Importing necessary libraries for data handling, computation, and model building.

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

## 📂 Dataset Preparation (Manual Implementation)  

Creating a small dataset to understand KNN step-by-step.

```python
data=pd.DataFrame({
    'House': ['H1', 'H2', 'H3', 'H4', 'H5', 'Test'],
    'Area': [2500, 3000, 1500, 1000, 2000, 1800],
    'Bedrooms': [3, 4, 2, 1, 3, 2],
    'Age': [5, 2, 20, 25, 10, 15],
    'Class': ['Expensive', 'Expensive', 'Affordable', 'Affordable', 'Affordable', '?']
})
```

## 📏 Distance Calculation  

Using Manhattan distance to measure similarity between data points.

## 🎯 Finding K Nearest Neighbors  

Sorting distances and selecting top K closest points.

```python
X=data[data['Class']!="?"]
y=data[data['Class']=="?"].iloc[0]
def manhattan_distance(p1,p2):
    return np.sum(np.abs(p1-p2))
distances=[]
for _,row in X.iterrows():
    dist=manhattan_distance(np.array([row['Area'], row['Bedrooms'], row['Age']]),
        np.array([y['Area'], y['Bedrooms'], y['Age']]))
    distances.append({"House":row['House'],"dist":dist,"Class":row['Class']})
dd=pd.DataFrame(distances)
print(dd)
k=3
dd_man=dd.sort_values(by='dist').head(k)
pred_class=dd_man['Class'].mode()[0]
pred_class
```

## 🧮 Prediction (Manual KNN)  

Using majority voting (mode) to determine the final class.

### 🔍 Observation:
The predicted class is determined based on the majority class among nearest neighbors.

## 🤖 KNN using Scikit-learn  

Applying KNN classifier using different distance metrics.

```python
from sklearn.neighbors import KNeighborsClassifier
train=pd.read_csv("knn.csv")
test=pd.read_csv("knn_test.csv")
X_train=train[['Weight','Height']]
y_train=train['Class']
X_test=test[['Weight','Height']]
```

## 📐 Distance Metrics Comparison  

Using different metrics:
- Manhattan  
- Squared Euclidean  
- Cosine  
- Minkowski  

```python
knn1=KNeighborsClassifier(n_neighbors=3,metric='manhattan')
knn2=KNeighborsClassifier(n_neighbors=3,metric='sqeuclidean')
knn3=KNeighborsClassifier(n_neighbors=3,metric='cosine')
knn4=KNeighborsClassifier(n_neighbors=3,metric='minkowski')

knn1.fit(X_train,y_train)
knn2.fit(X_train,y_train)
knn3.fit(X_train,y_train)
knn4.fit(X_train,y_train)
y_pred1=knn1.predict(X_test)
y_pred2=knn2.predict(X_test)
y_pred3=knn3.predict(X_test)
y_pred4=knn4.predict(X_test)
df_pred=pd.DataFrame({
    "Class1":y_pred1,"Class2":y_pred2,"Class3":y_pred3,"Class4":y_pred4
})
pred_val=df_pred.mode(axis=1)[0]
test['Class']=pred_val.values
test.to_csv("knn_test.csv",index=False)

```

## 🔄 Ensemble Prediction  

Combining predictions from multiple metrics using mode to improve robustness.

### 🔍 Observation:
Using multiple distance metrics reduces dependency on a single metric and improves prediction stability.

## 📊 Final Prediction Output  

Saving final predictions into dataset.

### 🔍 Final Insight:
KNN is a simple yet powerful algorithm that relies on distance metrics. Using multiple metrics and combining results improves reliability.
